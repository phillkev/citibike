```python
import pandas as pd
import numpy as np
from datetime import datetime

file = 'citibike/2018/2018-citibike-tripInfo-denorm-removeDate.csv'
testDF = pd.read_csv(file, encoding="UTF-8")
```

```python
testDF['startMonth'] = pd.to_datetime(testDF['startHour']).dt.strftime('%m')
testDF['Hour'] = pd.to_datetime(testDF['startHour']).dt.strftime('%H')
```

```python
MHGroupBy =  testDF.groupby(["startMonth","Hour"])

MHTripCount = MHGroupBy['tripduration'].count()
MHTripAvgDuration = MHGroupBy['tripduration'].mean()

MHSummary = pd.DataFrame ({"Trip Count": MHTripCount,
                              "Average Trip Duration": MHTripAvgDuration})

MHSummary.head()
```

```python
file = 'citibike/2018/2018-citibike-Month-hour_summary.csv'
MHSummary.to_csv(file, encoding="UTF-8")
```

```python
removeOutlier = testDF.loc[testDF["tripduration"] < 4500, :]

file = 'citibike/2018/2018-citibike-removed_GT4500.csv'
removeOutlier.to_csv(file, encoding="UTF-8")
```

```python
siteInfoDF = testDF.drop_duplicates(['start station id'])[['start station id']]
siteInfoDF.rename(columns={'start station id':'station_id'}, 
                 inplace=True)
file = 'citibike/2018/2018-citibike-siteInfo.csv'
siteInfoDF.to_csv(file,index=False)
```

```python
file = 'citibike/2018/2018-citibike-siteInfoend.csv'
endsiteinfo = pd.read_csv(file, encoding="UTF-8")
merge_table = pd.merge(removeOutlier, endsiteinfo, on="end station id")
```

```python
merge_table.head()
```

```python
onlyend = merge_table.loc[merge_table["sitetype"] == "Repair", :]
onlytrip = merge_table.loc[merge_table["sitetype"] == "Station", :]

file = 'citibike/2018/2018-citibike-removed_repair.csv'
onlytrip.to_csv(file, encoding="UTF-8")

file = 'citibike/2018/2018-citibike-only_repair.csv'
onlyend.to_csv(file, encoding="UTF-8")
```

```python
onlytrip.drop(['station_name', 'station_latitude', 'station_longitude', 'sitetype','startMonth','Hour'], axis=1, inplace=True)
file = 'citibike/2018/2018-citibike-removed_repair.csv'
onlytrip.to_csv(file, encoding="UTF-8")
```

```python
onlytrip = merge_table.loc[merge_table["sitetype"] == "Station", :]

file = 'citibike/2018/2018-citibike-removed_repair.csv'
onlytrip.to_csv(file, encoding="UTF-8")
```

```python
file = 'citibike/2018/2018-citibike-riderInfo.csv'
riderDF = pd.read_csv(file, encoding="UTF-8")

wuserandenssite = pd.merge(onlytrip, riderDF, on="id")

file = 'citibike/2018/2018-citibike-removed_repair.csv'
wuserandenssite.to_csv(file, encoding="UTF-8")
```

```python
wuserandenssite['age'] = 2018 - wuserandenssite['birth year']

```

```python
wuserandenssite.head()
```

```python
monthGroupBy =  wuserandenssite.groupby(["startMonth","Hour","usertype","gender","age"])

mTripCount = monthGroupBy['tripduration'].count()
mTripAvgDuration = monthGroupBy['tripduration'].mean()

MHSummary = pd.DataFrame ({"Trip Count": mTripCount,
                              "Average Trip Duration": mTripAvgDuration})

MHSummary.head()
```

```python
file = 'citibike/2018/2018-citibike-summary.csv'
MHSummary.to_csv(file, encoding="UTF-8")
```

```python
stationGroupBy =  wuserandenssite.groupby(["start station id","startMonth","Hour"])

sTripCount = stationGroupBy['tripduration'].count()
sTripAvgDuration = stationGroupBy['tripduration'].mean()

sSummary = pd.DataFrame ({"Trip Count": sTripCount,
                              "Average Trip Duration": sTripAvgDuration})

sSummary.head()

```

```python
file = 'citibike/2018/2018-citibike-station-summary.csv'
sSummary.to_csv(file, encoding="UTF-8")
```
